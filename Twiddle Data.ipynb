```python
ks = []
ms = []
rs = []
while True:
    c = input('New melody y/n: ')
    if c == 'n':
        break
    m = []
    k = input('Root of key: ')
    n = ''
    while n != 'q':
        n = input('Next number in melody/q: ')
        if n != 'q':
            m.append(n)
    r = []
    for i in range(len(m)):
        r.append(input('Next fraction of quarter note: '))
    print(k)
    print(m)
    print(r)
    ks.append(k)
    ms.append(m)
    rs.append(r)
```
