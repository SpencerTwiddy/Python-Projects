```python
def shiftsOfString(s):
    arr = []
    for i in range(len(s)):
        arr.append("")
        for j in range(len(s)):
            arr[i]+=(s[(i+j)%len(s)])
    return arr

def permsOfString(s):
    chars = set(s)
    split = []
    counts = []
    for i in range(len(chars)):
        counts.append(0)
    for i in s:
        for j in range(len(chars)):
            if i == list(chars)[j]:
                counts[j] += 1
    #print(chars)
    #print(counts)
    arr = [""]
    for ss in arr:
        for i in range(len(chars)):
            exs = ss + list(chars)[i]
            if not exs.count(list(chars)[i]) > counts[i]:
                arr.append(exs)
    #print(arr)
    #print(len(arr))
    answer = []
    for i in arr:
        if len(i)==7:
            answer.append(i)
    #print(answer)
    #print(len(answer))
    return answer

result = permsOfString("WWHWWWH")
allScales = set([])
palindromicScales = set([])
for i in result:
    arr = shiftsOfString(i)
    for j in arr:
        allScales.add(j)
        if j==j[::-1]:
            palindromicScales.add(j)
print(allScales)
print(len(allScales))
print(palindromicScales)
print(len(palindromicScales))
```
