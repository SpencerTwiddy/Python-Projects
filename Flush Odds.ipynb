```python
import random
```

```python
deck = []
for i in range(4):
    for j in range(13):
        if not (i==3 and j>=11):
            deck.append(str(i) + '.' + str(j))
print(deck)
```

```python
n = 10**5
s = 0
for i in range(n):
    d = deck.copy()
    h = []
    for j in range(5):
        k = random.randint(0, len(d) - 1)
        h.append(d[k])
        d.pop(k)
    c = 0
    for j in h:
        if j[0] == '3':
            c += 1
    if c >= 3:
        s += 1
print(s/n)
```
