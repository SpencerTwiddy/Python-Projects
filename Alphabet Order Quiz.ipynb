```python
import random
s = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
```

```python
g = ""
while len(g) < 3:
    r = random.randint(1,26)
    g = input(f"Letter {r}: ")
    print(s[r-1], end=" ")
```

```python
g = ""
while len(g) < 3:
    r = random.randint(1,26)
    g = input(f"{s[r-1]} is number: ")
    print(r, end=" ")
```

```python
g = ""
while len(g) < 3:
    m = random.getrandbits(1)
    if m:
        r = random.randint(1,26)
        g = input(f"Letter {r}: ")
        print(s[r-1], end=" ")
    else:
        r = random.randint(1,26)
        g = input(f"{s[r-1]} is number: ")
        print(r, end=" ")
```
