```python
# converts list arg and an optional separator into a one-line string to be printed
# list arg is list of strings preceded by optional ints denoting col width
# if col width is provided, the following string can be any type as it will be cast to str
# example: ['Usernames:', 8, 'Placid', 'Ethan', 8, 'Gravidus25', 'Spencer'] --> 'Usernames: Placid   Ethan Gravidus Spencer'
# if argument exceeds column size, the end will be cut off
def tabs(words, s = ' '):
    columns = []
    int_possible = True
    for e in words:
        if int_possible:
            if isinstance(e, int):
                col_max = e
                int_possible = False
            else:
                col_max = float('inf')
                columns.append(e)
        else:
            e = str(e)
            word = [' '] * col_max
            for i in range(col_max):
                if i == len(e): break
                word[i] = e[i]
            columns.append(''.join(word))
            int_possible = True
    return s.join(columns)

# converts a list of ints representing section lengths to a list of ints representing when each section starts
def lengths_to_starts(lengths):
    starts = [0]
    for i in range(len(lengths) - 1):
        starts.append(starts[i] + lengths[i])
    return starts
```
