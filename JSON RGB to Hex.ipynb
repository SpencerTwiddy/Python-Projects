```python
import pyperclip
```

```python
inp = '''
0: 240
1: 183
2: 25
'''
threeStr = inp.strip().split('\n')
threeHex = []
for i in threeStr:
    app = hex(int(i.split(':')[1].strip()))[2:]
    while len(app) < 2:
        app = '0' + app
    threeHex.append(app)
opts = ['', '#']
pound = opts[0]
final = pound + ''.join(threeHex).upper()
pyperclip.copy(final)
print(final)
```
