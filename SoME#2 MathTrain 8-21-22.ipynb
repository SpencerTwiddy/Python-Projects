```python
import math
```

```python
def mass(f, c, d): # f: functions for mass in each dimension, c: coefficients, d: dimensions, m: mass
    m = ''
    if len(f) == 1:
        return m + ' ' + point()
    m += str(c[-1]) + f[-1](d)
    return m + ' ' + mass(f[:-1], c[:-1], d) # average/default mass on boundary

def point():
    return '1'

def line(d):
    return f'{d}w^1'

def square(d):
    return f'{d**2}w^2'

def cube(d):
    return f'{d**3}w^3'

def circle(d):
    return f'{math.pi * d**2}w^2'

def sphere(d):
    return f'{4/3 * math.pi * d**3}w^3'

def rect(d):
    return f'{d[0] * d[1]}w^2'
```

```python
print(mass([point, line, square, cube], [1, 4, 6], 3))
print(mass([point, line, circle, sphere], [1, 2 * 3.14, 4 * 3.14], 3))
print(mass())
```
