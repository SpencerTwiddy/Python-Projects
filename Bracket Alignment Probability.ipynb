```python
def good_bracket(b):
    return five_per_side(b) and opposite_yourself(b)

def five_per_side(b):
    c = 0
    for i in b[:8]:
        if not_CPU(i):
            c += 1
    return c == 5

def not_CPU(p):
    return p >= 3 and p <= 12

def opposite_yourself(b):
    for i in b[:8]:
        if 15-i in b[:8] and not_CPU(i):
            return False
    return True
```
