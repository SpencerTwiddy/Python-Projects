```python
def tabs(words, s = ' '):
    columns = []
    int_possible = True
    for e in words:
        if int_possible:
            if isinstance(e, int):
                col_max = e
                int_possible = False
            else:
                col_max = float('inf')
                columns.append(e)
        else:
            e = str(e)
            word = [' '] * col_max
            for i in range(col_max):
                if i == len(e): break
                word[i] = e[i]
            columns.append(''.join(word))
            int_possible = True
    return s.join(columns)
```

```python
stop_term = 50
for i in range(stop_term):
    term = (i+1)**i
    print(tabs(['i:', 2, i, ' len:', 2, len(str(term)), ' ratio/i:', 5, (i+2)**(i+1)/term/i if i > 0 else 'N/A', ' term:', str(term)]))
```

```python
i = 10**6 # 1 million
term = (i+1)**i
print((i+2)**(i+1)/term/i) # ratio/i
```

```python
# ratio/i approaches e
# (i+2)**(i+1)/((i+1)**i)/i = ((i+2)*(1/(i+1)+1)^i)/i if i > 0
# the limit of this second form as i approaches infinity is equal to the limit of (1 + 1/i)^i, which is e
```
