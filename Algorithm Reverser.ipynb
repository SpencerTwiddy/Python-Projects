```python
def inverse(string):
    if not isValidAlg(string):
        return "Invalid Alg"
    origOrderList = putMovesInList(string)
    newList = []
    for i in origOrderList[::-1]:
        if len(i) == 1:
            newList.append(i+'\'')
        elif i[1] == '2':
            newList.append(i)
        else:
            newList.append(i[0])
    ans = ''
    for i in newList:
        ans += i + " "
    return ans[:-1]
    
def isValidAlg(string):
    string = string.replace(" ", "")
    nonDirs = ['F', 'R', 'L', 'U', 'D', 'B', 'M', 'E', 'S', 'f', 'r', 'l', 'u', 'd', 'b', 'x', 'y', 'z']
    dirs = ["'", "’", "2"]
    canAcceptDir = False
    for i in range(len(string)):
        if not canAcceptDir:
            if string[i] not in nonDirs:
                return False
            canAcceptDir = True
        else:
            if string[i] in dirs:
                canAcceptDir = False
            if string[i] not in nonDirs + dirs:
                return False
    return True

def putMovesInList(string):
    # only to be used on Valid Algs
    ans = []
    string = string.replace(" ", "")
    dirs = ["'", "’", "2"]
    currentMove = -1
    for i in range(len(string)):
        if string[i] in dirs:
            ans[currentMove] += string[i]
        else:
            ans.append(string[i])
            currentMove += 1
    return ans
```

```python
tests = ["R2     M d U R' U'"]
for i in tests:
    print("Input: " + f"{i}")
    print("Valid? " + f"{isValidAlg(i)}")
    print("List: " + f"{putMovesInList(i)}")
    print("Inverse: " + f"{inverse(i)}")
    print("")
```
