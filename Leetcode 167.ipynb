```python
# example
# input: [[1, 2, 7, 11, 15], 9]
# output: [2, 3]

def twoSum(inp):
    arr = inp[0]
    tgt = inp[1]
    comp = []
    for i in range(len(arr)):
        for j in range(len(comp)):
            if arr[i] == comp[j]:
                return [j + 1, i + 1]
        comp.append(tgt - arr[i])
    return []
```

```python
print(twoSum([[1, 2, 7, 11, 15], 9]))
print(twoSum([[1, 3, 4, 5, 7, 10, 11], 9]))
```

```python
def twoSumPointers(inp):
    arr = inp[0]
    tgt = inp[1]
    l = 0
    r = len(arr) - 1
    while l < r:
        if arr[l] + arr[r] > tgt:
            r -= 1
        elif arr[l] + arr[r] < tgt:
            l += 1
        else:
            return [l + 1, r + 1]
    return []
```

```python
print(twoSumPointers([[1, 2, 7, 11, 15], 9]))
print(twoSumPointers([[1, 3, 4, 5, 7, 10, 11], 9]))
```
