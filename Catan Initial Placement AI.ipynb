```python
import random
import logging
logging.basicConfig(level=logging.ERROR)

def show_steps():
    logging.getLogger().setLevel(logging.INFO)
    return

def hide_steps():
    logging.getLogger().setLevel(logging.ERROR)
    return

import math
```

```python
noh = 19 # number of hexes
pgd = 6 # polygon dimension
resource_types = ['ore', 'wheat', 'sheep', 'wood', 'brick']
max_number = 12
min_number = 2
resources = [''] * noh
for i in range(noh):
    resources[i] = resource_types[random.randint(0, len(resource_types) - 1)]
numbers = [0] * noh
for i in range(noh):
    numbers[i] = random.randint(min_number, max_number)
class Hex:
    def __init__(self, index, resource, number):
        self.index = index
        self.resource = resource
        self.number = number
        self.settlement_spots = [False] * pgd
        self.road_spots = [False] * pgd

class Settlement_spot:
    def __init__(self, hexes, port, occupancy):
        self.hexes = hexes
        self.port = port
        self.occupancy = occupancy
    
class Road_spot:
    def __init__(self, hexes, occupancy):
        self.hexes = hexes
        self.occupancy = occupancy
```

```python
def dots(h): # takes a Hex as param
    return 6 - abs(7 - h.number)
```

```python
hexes = []
for i in range(noh):
    hexes.append(Hex(i, resources[i], numbers[i]))
```
