```python
def addDigit(string, pool):
    ans = []
    for i in pool:
        ans.append(string + i)
    return ans

def perms(pool, n):
    ans = [""]
    for i in range(n):
        adds = []
        for j in ans:
            adds = adds + addDigit(j, pool)
        ans = adds
    return ans
```

```python
def waysQuantityOfTermsWithinBoundSumToTotal(quantity, maxNumber, total):
    ans = []
    # base exit if can't recurse
    if quantity <= 1:
        if total > maxNumber or total < 1:
            return []
        logging.info(f"RETURN: {total}")
        return [[total]]
    # recursive call
    big = min(total - 1, maxNumber)
    for i in range(big, 0, -1):
        logging.info(f"GOAL: {quantity} numbers <= {maxNumber} sum to {total}. Try biggest is {i}.")
        withoutI = waysQuantityOfTermsWithinBoundSumToTotal(quantity - 1, i - 1, total - i)
        bestPossible = 0
        for j in range(i, i-quantity, -1):
            bestPossible += j
        logging.debug(f"bestPossible: {bestPossible}")
        if len(withoutI) == 0 and bestPossible < total:
            logging.info(f"Impossible. Need number > {i} for {quantity} numbers to sum to {total}")
            break
        for j in withoutI:
            j.append(i)
            if len(j) == quantity:
                ans.append(sorted(j)[::-1])
    logging.info(f"RETURN: {ans}")
    return ans
```

```python
# Quiz

g = ""
b = 9
while len(g) < 3:
    q = random.randint(2, 7)
    lower = q * (q + 1) / 2
    higher = 10 * q - lower
    t = random.randint(lower, higher)
    g = input(f"How many ways can {q} unique positive integers <= {b} sum to {t}? Guess: ")
    w = waysQuantityOfTermsWithinBoundSumToTotal(q, b, t)
    l = len(w)
    print(f"Answer: {l} {w}")
```

```python
# Output space mapped

b = 9
s = []
print("0, 0: 1 []")
for q in range(1, 10):
    s.append([])
    lower = int(q * (q + 1) / 2)
    higher = int(10 * q - lower)
    for t in range(lower, higher + 1):
        s[q-1].append(waysQuantityOfTermsWithinBoundSumToTotal(q, b, t))
for i in range(len(s)):
    for j in range(len(s[i])):
        string = ""
        l = 0
        for k in s[i][j]:
            string += str(k)
            l += 1
        print(f"{i + 1}, {sum(list(s[i][j][0]))}: {l} {string}")
```
