```python
import logging
logging.basicConfig(level=logging.ERROR)
```

```python
def generate_fib(n):
    if n == 1:
        return [1]
    if n == 2:
        return [1, 1]
    fib = generate_fib(n-1)
    fib.append(fib[len(fib)-1]+fib[len(fib)-2])
    return fib

def generate_sums_of_fib(n):
    fib = generate_fib(n)
    sums_of_fib = []
    running_sum = 0
    for i in fib:
        running_sum += i
        sums_of_fib.append(running_sum)
    return sums_of_fib

def next_fib(fibs):
    if len(fibs) <= 1:
        return 1
    logging.info(f"next_fib({fibs}) == {fibs[len(fibs)-1]} + next_fib({fibs[:len(fibs)-2]})")
    return fibs[len(fibs)-1] + next_fib(fibs[:len(fibs)-2])
```

```python
showSteps = False
logging.getLogger().setLevel(logging.ERROR)
if showSteps: logging.getLogger().setLevel(logging.INFO)
    
t = generate_fib(15)
print(t)
print(generate_sums_of_fib(15))
print(next_fib(t[:-1]))

logging.getLogger().setLevel(logging.ERROR)
```
