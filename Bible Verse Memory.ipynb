```python
import random
class Meme:
  def __init__(self, index, book, chapter, verse_list, text):
    self.index = index
    self.book = book
    self.chapter = chapter
    self.verse_list = verse_list
    self.text = text
  def add_part(self):
    return

def remove_punc(s):
    s = s.replace('‘s', 'qQqQ')
    s = s.replace('\'s', 'qQqQ')
    for c in unwanted_punc:
        if c in s:
            s = s.replace(c, ' ')
    s = s.replace('qQqQ', '\'s')
    return s

def lengths_to_starts(lengths):
    starts = [0]
    for i in range(len(lengths) - 1):
        starts.append(starts[i] + lengths[i])
    return starts

def tabs(words, s = ' '):
    columns = []
    int_possible = True
    for e in words:
        if int_possible:
            if isinstance(e, int):
                col_max = e
                int_possible = False
            else:
                col_max = float('inf')
                columns.append(e)
        else:
            word = [' '] * col_max
            for i in range(col_max):
                if i == len(e): break
                word[i] = e[i]
            columns.append(''.join(word))
            int_possible = True
    return s.join(columns)

def get_words(s):
    return remove_punc(s).split()
```
