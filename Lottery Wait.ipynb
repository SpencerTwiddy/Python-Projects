```python
import random
import numpy as np
from collections import Counter
import math
```

```python
n = 40000
waitTimes = []
invP = 36
for i in range(n):
    waiting = True
    att = 0
    while waiting:
        att+=1
        if random.randint(1,invP)==1:
            waiting = False
            waitTimes.append(att)
```

```python
print(np.average(np.array(waitTimes)))
# ~ invP
```
