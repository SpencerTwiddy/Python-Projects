```python
import cmath
import math
```

```python
def printi(imaginary_number):
    s = str(round(imaginary_number.real,2)) + " + " + str(round(imaginary_number.imag,2)) + "i"
    print(s)
    return s
```

```python
r = 1
ph = 2*math.pi/5
z1 = cmath.rect(r, ph)
z2 = z1 ** 2
z3 = z1 ** 3
z4 = z1 ** 4
z5 = z1 ** 5
printi(z1)
printi(z2)
printi(z3)
printi(z4)
printi(z5)
```
