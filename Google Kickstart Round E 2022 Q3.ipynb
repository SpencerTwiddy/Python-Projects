```python
inp = [3, 4, 'abba', 4, 'cccc', 6, 'cdccdc']
```

```python
def solve(length, pal):
    cutoff = 0
    ans = pal[::-1]
    while True:
        if abs(cutoff) >= len(pal):
            break
        if cutoff == 0:
            app = ans
        else:
            app = pal[::-1][:cutoff]
        maybe = pal + app
        palin = True
        for i in range(len(maybe)//2+1):
            if maybe[i] != maybe[-i-1]:
                palin = False
        if palin:
            ans = app
        cutoff -= 1
    return ans
        

lens = []
pals = []
for i in range(inp[0]):
    length = inp[2*i+1]
    pal = inp[2*i+2]
    print(solve(length, pal))
    
```
