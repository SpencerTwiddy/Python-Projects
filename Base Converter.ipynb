```python
def convertBase(req, inpV, outpV):
    req = str(req)
    nreq = []
    for i in req:
        try:
            nreq.append(int(i))
        except:
            nreq.append(ord(i)-55)
    place = len(nreq)
    decEquivs = []
    for d in nreq:
        place -= 1
        decEquivs.append(int(d)*inpV**place)
    tot = sum(decEquivs)
    place = 0
    while outpV ** place <= tot:
        place += 1
    ans = ""
    while place > 0:
        place -= 1
        val = (tot - tot % outpV**place) / outpV**place
        if len(str(int(val)))==1:
            appendix = str(int(val))
        else:
            appendix = chr(55+int(val))
        ans += appendix
        tot = tot - val * outpV**place
    return ans
```

```python
choices = {"HEX": 16, "DEC": 10, "OCT": 8, "SEX": 6, "BIN": 2, "Other": -1}
for i in range(len(choices)):
    print(f"{i}) {list(choices)[i]}")
inp = -1
while int(inp) not in range(len(choices)):
    inp = input("From: ")
    try:
        int(inp)
    except:
        inp = -1
outp = -1
inp = int(inp)
if list(choices)[inp] == "Other":
    invalidInpV = True
    while invalidInpV:
        inpV = input("Other base: ")
        try:
            inpV = int(inpV)
            if inpV>1 and inpV<37:
                invalidInpV = False
        except:
            invalidInpV = True
    base1 = "Base " + str(inpV)
else:
    base1 = list(choices)[inp]
    inpV = choices[base1]
while int(outp) not in range(len(choices)):
    outp = input("To: ")
    try:
        int(outp)
    except:
        outp = -1
outp = int(outp)
if list(choices)[outp] == "Other":
    invalidOutpV = True
    while invalidOutpV:
        outpV = input("Other base: ")
        try:
            outpV = int(outpV)
            if outpV>1 and outpV<37:
                invalidOutpV = False
        except:
            invalidOutpV = True
    base2 = "Base " + str(outpV)
else:
    base2 = list(choices)[outp]
    outpV = choices[base2]
print(f"Converting from {base1} to {base2}")
a = list(range(48,57+1))+list(range(65,90+1))
invalidReq = True
while invalidReq:
    req = input(f"{base1}: ")
    invalidReq = False
    for i in req:
        if ord(i) not in a:
            invalidReq = True
print(f"{base2}: {convertBase(req, inpV, outpV)}")
```
