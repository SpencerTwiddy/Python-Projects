```python
def sortChars(string):
    codes = {}
    for i, v in enumerate(string):
        codes[str(i)] = ord(v)
    # https://stackoverflow.com/questions/613183/how-do-i-sort-a-dictionary-by-value
    newDict = {k: v for k, v in sorted(codes.items(), key=lambda item: item[1])}
    res = ''
    for i in newDict.keys():
        res += chr(codes[i])
    return res

def tabs(words, s = ' '):
    columns = []
    int_possible = True
    for e in words:
        if int_possible:
            if isinstance(e, int):
                col_max = e
                int_possible = False
            else:
                col_max = float('inf')
                columns.append(e)
        else:
            e = str(e)
            word = [' '] * col_max
            for i in range(col_max):
                if i == len(e): break
                word[i] = e[i]
            columns.append(''.join(word))
            int_possible = True
    return s.join(columns)

testStrs = ['hello world', 'how many calls', 'very uneventful', 'abcdefghijklmnopqrstuvwxyz']
for i in testStrs:
    x = sortChars(i.replace(' ', ''))
    y = "".join(set(x))
    print(tabs(['Full:', 40, x + ' (' + str(len(x)) + ')', 'Unique:', y + ' (' + str(len(y)) + ')']))
```
