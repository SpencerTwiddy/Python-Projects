```python
from __future__ import print_function
import os
import re


__ALL__ = [ 'colored', 'cprint' ]

VERSION = (1, 1, 0)

ATTRIBUTES = dict(
        list(zip([
            'bold',
            'dark',
            '',
            'underline',
            'blink',
            '',
            'reverse',
            'concealed'
            ],
            list(range(1, 9))
            ))
        )
del ATTRIBUTES['']

ATTRIBUTES_RE = '\033\[(?:%s)m' % '|'.join(['%d' % v for v in ATTRIBUTES.values()])

HIGHLIGHTS = dict(
        list(zip([
            'on_grey',
            'on_red',
            'on_green',
            'on_yellow',
            'on_blue',
            'on_magenta',
            'on_cyan',
            'on_white'
            ],
            list(range(40, 48))
            ))
        )

HIGHLIGHTS_RE = '\033\[(?:%s)m' % '|'.join(['%d' % v for v in HIGHLIGHTS.values()])

COLORS = dict(
        list(zip([
            'grey',
            'red',
            'green',
            'yellow',
            'blue',
            'magenta',
            'cyan',
            'white',
            ],
            list(range(30, 38))
            ))
        )

COLORS_RE = '\033\[(?:%s)m' % '|'.join(['%d' % v for v in COLORS.values()])

RESET = '\033[0m'
RESET_RE = '\033\[0m'


def colored(text, color=None, on_color=None, attrs=None):
    """Colorize text, while stripping nested ANSI color sequences.
    Available text colors:
        red, green, yellow, blue, magenta, cyan, white.
    Available text highlights:
        on_red, on_green, on_yellow, on_blue, on_magenta, on_cyan, on_white.
    Available attributes:
        bold, dark, underline, blink, reverse, concealed.
    Example:
        colored('Hello, World!', 'red', 'on_grey', ['blue', 'blink'])
        colored('Hello, World!', 'green')
    """
    if os.getenv('ANSI_COLORS_DISABLED') is None:
        fmt_str = '\033[%dm%s'
        if color is not None:
            text = re.sub(COLORS_RE + '(.*?)' + RESET_RE, r'\1', text)
            text = fmt_str % (COLORS[color], text)
        if on_color is not None:
            text = re.sub(HIGHLIGHTS_RE + '(.*?)' + RESET_RE, r'\1', text)
            text = fmt_str % (HIGHLIGHTS[on_color], text)
        if attrs is not None:
            text = re.sub(ATTRIBUTES_RE + '(.*?)' + RESET_RE, r'\1', text)
            for attr in attrs:
                text = fmt_str % (ATTRIBUTES[attr], text)
        return text + RESET
    else:
        return text


def cprint(text, color=None, on_color=None, attrs=None, **kwargs):
    """Print colorize text.
    It accepts arguments of print function.
    """

    print((colored(text, color, on_color, attrs)), **kwargs)


if __name__ == '__main__':
    print('Current terminal type: %s' % os.getenv('TERM'))
    print('Test basic colors:')
    cprint('Grey color', 'grey')
    cprint('Red color', 'red')
    cprint('Green color', 'green')
    cprint('Yellow color', 'yellow')
    cprint('Blue color', 'blue')
    cprint('Magenta color', 'magenta')
    cprint('Cyan color', 'cyan')
    cprint('White color', 'white')
    print(('-' * 78))

    print('Test highlights:')
    cprint('On grey color', on_color='on_grey')
    cprint('On red color', on_color='on_red')
    cprint('On green color', on_color='on_green')
    cprint('On yellow color', on_color='on_yellow')
    cprint('On blue color', on_color='on_blue')
    cprint('On magenta color', on_color='on_magenta')
    cprint('On cyan color', on_color='on_cyan')
    cprint('On white color', color='grey', on_color='on_white')
    print('-' * 78)

    print('Test attributes:')
    cprint('Bold grey color', 'grey', attrs=['bold'])
    cprint('Dark red color', 'red', attrs=['dark'])
    cprint('Underline green color', 'green', attrs=['underline'])
    cprint('Blink yellow color', 'yellow', attrs=['blink'])
    cprint('Reversed blue color', 'blue', attrs=['reverse'])
    cprint('Concealed Magenta color', 'magenta', attrs=['concealed'])
    cprint('Bold underline reverse cyan color', 'cyan',
            attrs=['bold', 'underline', 'reverse'])
    cprint('Dark blink concealed white color', 'white',
            attrs=['dark', 'blink', 'concealed'])
    print(('-' * 78))

    print('Test mixing:')
    cprint('Underline red on grey color', 'red', 'on_grey',
            ['underline'])
    cprint('Reversed green on red color', 'green', 'on_red', ['reverse'])
```

```python
def best_continuations(dictionary, words, colors, search_how_many, show_how_many):
    
    if len(words) != len(colors):
        print("len(words) != len(colors)")
        return
    search_how_many = min(len(dictionary), search_how_many)
    slow_part_first_time = True
    
    def possible_given_input(dictionary, words, colors):
        # regex match every word in dictionary to form the answer might take
        # assumes word is 5 capital letters
        minimum_occurrences = {}
        for letter in alphabet:
            minimum_occurrences[letter] = 0
        bracket_insides = []
        for letter in range(5):
            bracket_insides.append('^0')
        for clue in range(len(words)):
            for pos in range(5):
                match = colors[clue][pos]
                if match == '2':
                    bracket_insides[pos] = words[clue][pos]
                if match == '1':
                    if bracket_insides[pos][0] == '^':
                        bracket_insides[pos] += words[clue][pos]
                if match == '0':
                    for pos2 in range(5):
                        if bracket_insides[pos2][0] == '^':
                            bracket_insides[pos2] += words[clue][pos]
                # now weed out words that don't have letters we know they must
                letter_and_2_positions = [0] * 5
                for pos2 in range(5):
                    for clue2 in range(len(words)):
                        if colors[clue2][pos2] == "2" and words[clue2][pos2] == words[clue][pos]:
                            letter_and_2_positions[pos2] = 1
                has_1 = 0
                for pos2 in range(5):
                    if words[clue][pos2] == words[clue][pos] and colors[clue][pos2] == "1":
                        if words[clue].count('2') == sum(letter_and_2_positions):
                            has_1 = 1
                letter_min = sum(letter_and_2_positions) + has_1
                minimum_occurrences[words[clue][pos]] = max(letter_min, minimum_occurrences[words[clue][pos]])
        pattern = ''
        for inside in bracket_insides:
            pretty_inside = inside
            if '^' in inside:
                pretty_inside = f"[^{inside[2:]}]"
            pattern += pretty_inside
        # first create pool of words that conform to first 2 constraints (minimum and regex)
        pool_first_2 = []
        for potential_word in dictionary:
            if not re.match(pattern, potential_word) == None:
                possible = True
                for letter in alphabet:
                    if potential_word.count(letter) < minimum_occurrences[letter]:
                        possible = False
                if possible:
                    pool_first_2.append(potential_word)
        # now make third constraint group by replacing '2' with
        # every letter combo that overlaps
#         results = []
#         group_duplicates = patterns_given_emoji_grids(emoji_grids)
#         progress = 0
#         loops = min(9, len(pool_first_2))
#         waiting = loops
#         if len(emoji_grids) > 0 and slow_part_first_time:
#             print(f"Waiting: {waiting}", end = "")
#         for potential_word in pool_first_2:
#             if len(emoji_grids) > 0 and slow_part_first_time:
#                 progress += 1
#                 if progress / len(pool_first_2) > (loops + 1 - waiting) / loops:
#                     waiting -= 1
#                     print(waiting, end = "")
#             if possible_given_emoji_grids(potential_word, group_duplicates, globals()['dictionary']):
#                 results.append(potential_word)
#         if len(emoji_grids) > 0 and slow_part_first_time:
#             print()
        return [pool_first_2, minimum_occurrences, pattern]
    
    def repeat_metric(word):
        # returns how many of non-confirmed letters are repeats
        # the lower the better
        unique_letters_in_open_spots = set([])
        twos_accumulated = [0] * 5
        for clue in range(len(words)):
            for pos in range(5):
                if colors[clue][pos] == "2":
                    twos_accumulated[pos] = 1
        for pos in range(len(twos_accumulated)):
            if twos_accumulated[pos] == 0:
                unique_letters_in_open_spots.add(word[pos])
        return (5-sum(twos_accumulated)) - len(unique_letters_in_open_spots)
    
    # taken from stackoverflow on "How can you delete similar characters at the same positions in 2 strings"
    def compare(a, b):
        s = zip(*[i for i in zip_longest(a, b, fillvalue = "") if i[0]!=i[1]])
        return [''.join(k) for k in s]
    
    def display_list_of_items(items, color):
        # where an item means something that when converted to a string
        # will contain a 5-letter word beginning at the 3rd character
        # the first word will be emphasized with the given color
        if len(items) == 0:
            return ''
        display_str = f"{items[:show_how_many]}"[1:-1]
        if len(items) > 0:
            comma = ''
            if len(items) > 1:
                comma = ', '
            post_word_part = f"{items[0]}"
            pre_word_part = post_word_part[:2]
            word_part = post_word_part[2:7]
            post_word_part = post_word_part[7:]
            display_str = pre_word_part + colored(f"{word_part}", color, attrs=['bold', 'underline']) + post_word_part + comma + f"{items[1:show_how_many]}"[1:-1]
        return display_str
    
    display_wordle(words, colors)
    
    # Split up pool by repeat metric
    if len(words) > 0:
        pgi = possible_given_input(dictionary, words, colors)
        pool = pgi[0]
        print(f"Minimum occurrences: {pgi[1]}\nAnswer pattern: {pgi[2]}")
    else:
        pool = dictionary
    scores = [0] * len(pool)
    repeat_metrics = [0] * len(pool)
    for pool_idx in range(len(pool)):
        for pos in range(5):
            letter = pool[pool_idx][pos]
            scores[pool_idx] += counts[pos][letter]
            repeat_metrics[pool_idx] = repeat_metric(pool[pool_idx])
    pool_score_maps = [{}, {}, {}, {}, {}]
    for pool_idx in range(len(pool)):
        pool_score_maps[repeat_metrics[pool_idx]][pool[pool_idx]] = scores[pool_idx]    
    ranked_pools = [{}, {}, {}, {}, {}]
    for rep_met in range(len(ranked_pools)):
        ranked_pools[rep_met] = {k: v for k, v in sorted(pool_score_maps[rep_met].items(), key=lambda item: item[1])[::-1]}
    for display_line in range(len(ranked_pools)):
        display_list = display_list_of_items(list(ranked_pools[display_line].items()), 'cyan')
        if len(display_list) > 0:
            print(f"{min(len(list(ranked_pools[display_line].items())), show_how_many)}/{len(list(ranked_pools[display_line].items()))} with {display_line} repeats: {display_list}")
    
    # Common Words
    more_likely = []
    for display_line in range(len(ranked_pools)):
        for word in list(ranked_pools[display_line].items()):
            if word[0] in common:
                more_likely.append(word)
    display_list = display_list_of_items(more_likely, 'blue')
    print(f"Common words remaining: {display_list}")
    
    def share_3_letters(word1, word2):
        s1 = set(word1)
        s2 = set(word2)
        return abs(len(s1-s2) - len(s1)) >= 2            
    
    def most_left(cand, words_to_cover):
        words_plus_potential = words + [cand]
        return_value = 0
        for decimal in range(3**5):
            colors_plus_potential = str(np.base_repr(decimal, base=3))
            while len(colors_plus_potential) < 5:
                colors_plus_potential = "0" + colors_plus_potential
            colors_plus_potential = colors + [colors_plus_potential]
            new_return_value = len(possible_given_input(words_to_cover, words_plus_potential, colors_plus_potential)[0])
            return_value = max(new_return_value, return_value)
        if return_value == 0:
            if cand in words_to_cover:
                return_value = 1
        return return_value
        
    # Differentiating letters
    off_by_one_letters = {}
    chances_to_be_off = 0
    actually_off = 0
    for letter in alphabet:
        off_by_one_letters[letter] = 0
    for entry in range(len(ranked_pools[0])):
        for entry2 in range(len(ranked_pools[0])):
            if entry != entry2:
                letters_off = compare(list(ranked_pools[0].keys())[entry], list(ranked_pools[0].keys())[entry2])
                chances_to_be_off += 1
                if len(letters_off[0]) == 1:
                    actually_off += 1
                    for one_that_is_different in range(len(letters_off)):
                        off_by_one_letters[letters_off[one_that_is_different]] += 1
    sorted_off_by_one = {k: v for k, v in sorted(off_by_one_letters.items(), key=lambda item: item[1])[::-1]}
    letters_with_nonzero_off_by_one = []
    for letter in alphabet:
        if off_by_one_letters[letter] > 0:
            letters_with_nonzero_off_by_one.append(letter)
    percentage = "{:.2f}".format(100 * actually_off / (chances_to_be_off + 0.00001))
    #print(f"Off by one letter: {percentage}%  Differentiating letters: {sorted_off_by_one}")    
    
    # When letters are unique to a word but are not off by 1    
    unique_letter_tracker = {}
    for letter in alphabet:
        unique_letter_tracker[letter] = 0
    for pool_idx in range(len(pool)):
        for letter in pool[pool_idx]:
            unique_letter_tracker[letter] += 1
    for letter in alphabet:
        if unique_letter_tracker[letter] == 1 and letter not in letters_with_nonzero_off_by_one:
            letters_with_nonzero_off_by_one.append(letter)
    # letters_with_nonzero_off_by_one = ['F', 'K', 'D', 'T', 'G', 'U', 'Q'] # manual input
    
    # Covering words
    count_threshold = 5
    covering_words = []
    for random_shot in range(10):
        covering_words.append(globals()['dictionary'][len(globals()['dictionary']) // 10 * random_shot])
    while len(covering_words) <= search_how_many and count_threshold > 0:
        for possible_word in range(len(globals()['dictionary'])):
            count = 0
            for letter in letters_with_nonzero_off_by_one:
                if letter in globals()['dictionary'][possible_word]:
                    count += 1
            if count >= count_threshold:
                covering_words.append(globals()['dictionary'][possible_word])
        count_threshold -= 1
    scored_covering_words = {}
    covering_scores = [0] * len(covering_words)
    for covering_word in range(len(covering_words)):
        covering_word_no_repeats = str("".join(set(covering_words[covering_word])))
        for letter in covering_word_no_repeats:
            covering_scores[covering_word] += sorted_off_by_one[letter]
        scored_covering_words[covering_words[covering_word]] = covering_scores[covering_word]
    ranked_scored_covering_words = {k: v for k, v in sorted(scored_covering_words.items(), key=lambda item: item[1])[::-1]}
    display_list = display_list_of_items(list(ranked_scored_covering_words.items()), 'red')
    print(f"Covering depth: {count_threshold + 1}")
    
    # Sort top suggestions u hidden covering words by
    # how many options there will be at worst in the next step
    slow_part_first_time = False
    amounts_left = {}
    print(f"Wait 3: ", end = "")
    for rep in range(5):
        print(f"{5-rep}", end = "")
        for good_word_idx in range(min(search_how_many, len(ranked_pools[rep]))):
            good_word = list(ranked_pools[rep].keys())[good_word_idx]
            amounts_left[good_word] = most_left(good_word, pool)
    loops = min(9, min(search_how_many, len(more_likely)))
    waiting = loops
    print(f"\nWait 2: {waiting}", end = "")
    for good_word_idx in range(min(search_how_many, len(more_likely))):
        if good_word_idx / min(search_how_many, len(more_likely)) > (loops + 1 - waiting) / loops:
            waiting -= 1
            print(waiting, end = "")
        good_word = list(more_likely)[good_word_idx][0]
        amounts_left[good_word] = most_left(good_word, pool)
    loops = min(9, min(search_how_many, len(ranked_scored_covering_words)))
    waiting = loops
    print(f"\nWait 1: {waiting}", end = "")
    for good_word_idx in range(min(search_how_many, len(ranked_scored_covering_words))):
        if good_word_idx / min(search_how_many, len(ranked_scored_covering_words)) > (loops + 1 - waiting) / loops:
            waiting -= 1
            print(waiting, end = "")
        good_word = list(ranked_scored_covering_words.keys())[good_word_idx]
        amounts_left[good_word] = most_left(good_word, pool)
    print()
    ranked_amounts_left = {k: v for k, v in sorted(amounts_left.items(), key=lambda item: item[1])[::]}
    display_list = display_list_of_items(list(ranked_amounts_left.items()), 'red')
    print(f"Worst case at next step: {display_list}")
    return
```

```python
def two_bad_punc(char_pair):
    if '/' in char_pair:
        char_pair = char_pair.replace("/","")
        if char_pair in '+-/*':
            return True
    if '*' in char_pair:
        char_pair = char_pair.replace("*","")
        if char_pair in '+-/':
            return True
    return False

def valid_nerdle_start(start, pos):
    if '//' in start:
        return False
    if start.count('=') > 1:
        return False
    for pos in range(len(start)-1):
        if two_bad_punc(start[pos] + start[pos + 1]):
            return False
    return True
```

```python
def antiwordle(words, colors, choices, how_many):
    
    
    def repeat_metric(word):
        # returns how many of non-confirmed letters are repeats
        # the lower the better
        unique_letters_in_open_spots = set([])
        twos_accumulated = [0] * 5
        for clue in range(len(words)):
            for pos in range(5):
                if colors[clue][pos] == "2":
                    twos_accumulated[pos] = 1
        for pos in range(len(twos_accumulated)):
            if twos_accumulated[pos] == 0:
                unique_letters_in_open_spots.add(word[pos])
        return (5-sum(twos_accumulated)) - len(unique_letters_in_open_spots)
    
    
    
    pattern = ['','','','','']
    requirements = {}
    for letter in alphabet:
        requirements[letter] = 0
    absents = set([])
    for word in range(len(words)):
        for pos in range(len(words[word])):
            if colors[word][pos] == '0':
                absents.add(words[word][pos])
            if colors[word][pos] == '1':
                requirements[words[word][pos]] = 1 # doesn't work in all cases
            if colors[word][pos] == '2':
                pattern[pos] = words[word][pos]
    for col in range(len(pattern)):
        if len(pattern[col]) == 0:
            if len(absents) > 0:
                pattern[col] = '[^' + ''.join(absents) + ']'
            else:
                pattern[col] = '.'
    pattern = ''.join(pattern)
    print(pattern)
    print(requirements)
    pool = []
    for word in choices:
        bad_on_reqs = False
        for req in alphabet:
            if requirements[req] > 0 and not req in word:
                bad_on_reqs = True
                break
        if bad_on_reqs:
            continue
        if not re.match(pattern, word) == None:
            pool.append(word)
            
    scores = [0] * len(pool)
    repeat_metrics = [0] * len(pool)
    for pool_idx in range(len(pool)):
        for pos in range(5):
            letter = pool[pool_idx][pos]
            scores[pool_idx] += positional_z_score[pos][letter] + overall_z_scores[letter] * 2
            repeat_metrics[pool_idx] = repeat_metric(pool[pool_idx])
    pool_score_maps = [{}, {}, {}, {}, {}]
    for pool_idx in range(len(pool)):
        pool_score_maps[repeat_metrics[pool_idx]][pool[pool_idx]] = float("{:0.1f}".format(scores[pool_idx]))
    ranked_pools = [{}, {}, {}, {}, {}]
    for rep_met in range(len(ranked_pools)):
        ranked_pools[rep_met] = {k: v for k, v in sorted(pool_score_maps[rep_met].items(), key=lambda item: item[1])[::]}
    next_choices = []
    for repeat_val in ranked_pools:
        next_choices += repeat_val
        if len(list(repeat_val.items())[:how_many]) > 0:
            print(f"{list(repeat_val.items())[:how_many]} ({len(list(repeat_val.items())[:how_many])}/{len(list(repeat_val.items()))})")
    combined_rank = []
    counters = [0] * 5
    while sum(counters) < len(next_choices):
        best_score = 9999
        best_counter = 0
        for i in range(len(ranked_pools)):
            if len(list(ranked_pools[i].values())) > counters[i]:
                if list(ranked_pools[i].values())[counters[i]] < best_score:
                    best_score = list(ranked_pools[i].values())[counters[i]]
                    best_counter = i
#         print(f"Next: {best_counter}, ", end = "")
#         print(f"Iterations: {counters}, Word: {list(ranked_pools[best_counter].keys())[counters[best_counter]]}")
        if not list(ranked_pools[best_counter].keys())[counters[best_counter]] in common:
            combined_rank.append(list(ranked_pools[best_counter].keys())[counters[best_counter]])
        counters[best_counter] += 1
    print(f"By score, common words removed: {combined_rank[:how_many]}")
    return next_choices
```
