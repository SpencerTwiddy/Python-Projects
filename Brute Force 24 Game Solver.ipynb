```python
from itertools import permutations, product
```

```python
def intersplice(list0, list1, bool_list):
    if len(list0) + len(list1) != len(bool_list) or sum(bool_list) != len(list1):
        raise ValueError("list0 and list1 don't map to bool_list")
    ans = []
    c0 = 0
    c1 = 0
    for b in bool_list:
        if b:
            ans.append(list1[c1])
            c1 += 1
        else:
            ans.append(list0[c0])
            c0 += 1
    return ans
```

```python
def construct_splices(ans, splice, nums_available, ops_available):
    if nums_available < 1 and ops_available < 1:
        ans.append(splice)
    if nums_available >= 1:
        construct_splices(ans, splice + [False], nums_available - 1, ops_available + 1)
    if ops_available >= 1:
        construct_splices(ans, splice + [True], nums_available, ops_available - 1)
    return ans
```

```python
def generateSeqs(digits, ops):
    splices = construct_splices([], [], len(digits), -1)
    d = list(permutations(digits))
    o = list(product(ops, repeat = sum(splices[0])))
    ans = []
    for i in d:
        for j in o:
            for k in splices:
                ans.append(intersplice(i, j, k))
    return ans

def evaluate(seq, ops):
    stack = []
    for i in seq:
        if str(i) in ops:
            try:
                new_stack_val = eval(str(stack[-2]) + str(i) + str(stack[-1]))
            except ZeroDivisionError:
                return None
            stack.pop(-1)
            stack.pop(-1)
            stack.append(new_stack_val)
        else:
            stack.append(i)
    return stack[0]
```

```python
goal = 24
ops = '+-*/'
def solve(digits):
    seqs = generateSeqs(digits, ops)
    ans = []
    for seq in seqs:
        if evaluate(seq, ops) == goal:
            ans.append(seq)
    return ans
```

```python
def solToString(formal_solution):
    s = []
    for f in formal_solution:
        s.append(''.join([str(d) for d in f]))
    return s
```

```python
problems = [
    [2,2,5,7],
    '2257',
    '2885',
    '1245',
    '2269',
    '2653',
    '1449',
    '6623',
    [2,2,36,3],
    ['2', '2', '36', '3'],
    [24],
    [-24],
    '38',
    [2,60,-48],
    '12345' 
]

for p in problems:
    formal_solution = solve(p)
    print(formal_solution)
    print(solToString(formal_solution), '\n')
```
