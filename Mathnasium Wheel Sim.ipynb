```python
import random
```

```python
def roll(s):
    a = random.randint(1, 6)
    b = random.randint(1, 6)
    s += str(a) + str(b)
    if a+b != 7 and a+b != 11 and a != b:
        return [0, s]
    if a == b:
        t = roll('')
        return [1 + t[0], s + t[1]]
    return [1, s]
```

```python
n = 10**3
c = {}
ss = {}
for i in range(n):
    t = roll('')
    c[i] = t[0]
    ss[i] = t[1]
print(ss)
print(c)
print(sum(c.values())/n)
```

```python
# 1: 1
# 2: Double

def simple(s):
    a = random.randint(1, 2)
    s += str(a)
    if a == 2:
        t = simple('')
        return [2 * t[0], s + t[1]]
    return [1, s]
```

```python
# Wheel: 2 2s, 2 3s, 2 4s, 2 5s, 1 6, 1 jackpot, 1 double, 1 triple
wheel = [2,2,3,3,4,4,5,5,6,24,-2,-3]
def spin(s):
    i = random.randint(0, 11)
    a = wheel[i]
    s += str(a)
    if a < 0:
        t = spin('')
        return [-a * t[0], s + t[1]]
    return [a, s]
```

```python
n = 10**1
c = {}
ss = {}
for i in range(n):
    t = simple('')
    c[i] = t[0]
    ss[i] = t[1]
print(ss)
print(c)
print(sum(c.values())/n)
```
