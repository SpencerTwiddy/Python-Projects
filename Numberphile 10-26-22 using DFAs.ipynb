```python
def tabs(words, s = ' '):
    columns = []
    int_possible = True
    for e in words:
        if int_possible:
            if isinstance(e, int):
                col_max = e
                int_possible = False
            else:
                col_max = float('inf')
                columns.append(e)
        else:
            e = str(e)
            word = [' '] * col_max
            for i in range(col_max):
                if i == len(e): break
                word[i] = e[i]
            columns.append(''.join(word))
            int_possible = True
    return s.join(columns)
```

```python
import logging
logging.basicConfig(level=logging.ERROR)
```

```python
def statesEqual(s1, s2):
    return s1['name'] == s2['name']
```

```python
def checkAcceptance(s, DFA):
    logging.info(f"s: {s}")
    states = DFA[0]
    arrows = DFA[1]
    point = -1
    for i, state in enumerate(states):
        if state['initial']:
            point = i
            break
    logging.info(f'point: {point}')
    for i, ch in enumerate(s):
        for j, arrow in enumerate(arrows):
            if statesEqual(states[point], arrow['fromNode']) and arrow['symbol'] == int(ch):
                point = int(arrow['toNode']['name'][-1])
                logging.info(f"point: {point}")
                break
    logging.info(f"states[point]['accepting']: {states[point]['accepting']}")
    return states[point]['accepting']
```

```python
# Runs

DFAs = {'4': DFA4, '5': DFA5} # Include all DFAs created in p5js here

for k, DFA in DFAs.items():
    n = int(k)
    print(f'n = {n}')
    tests = []
    for i in range(1, n + 1):
        for j in range(2**i, 2**(i+1)):
            tests.append(bin(j)[3:])
    c = 0
    l = []
    for testString in tests:
        showSteps = False # Set to True for debugging
        logging.getLogger().setLevel(logging.ERROR)
        if showSteps: logging.getLogger().setLevel(logging.INFO)

        result = checkAcceptance(testString, DFA)

        logging.getLogger().setLevel(logging.ERROR)

        print(tabs(['String:', n, testString, '  Accepted?', str(result)]))
        if result:
            c += 1
            l.append(testString)
    print(f'Number accepted: {c}')
    print(f'Strings accepted: {l}')
    l3 = []
    for i in l:
        l3.append(int(i, 2))
    print(f'In decimal: {l3}')
    l2 = sorted(l3)
    print(f'Sorted: {l2}')
```
