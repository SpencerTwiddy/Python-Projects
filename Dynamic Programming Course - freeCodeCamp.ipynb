```python
def fib_naive(n):
    if n < 2: return 1
    return fib_naive(n - 2) + fib_naive(n - 1)

def fib_memo(n, m = None):
    if not m:
        m = [-1] * n
    if n < 2: return 1
    if m[n - 2] < 0:
        m[n - 2] = fib_memo(n - 2, m)
    if m[n - 1] < 0:
        m[n - 1] = fib_memo(n - 1, m)
    return m[n - 2] + m[n - 1]
```

```python
fib_naive(33)
```
