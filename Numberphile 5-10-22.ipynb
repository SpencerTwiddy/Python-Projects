```python
import random
```

```python
def random_spot_on_cake():
    return random.random()

def between(mid, x, y):
    return mid > min(x, y) and mid < max(x, y)
```

```python
trials = 999999
successes = 0
for trial in range(trials):
    candle1 = random_spot_on_cake()
    candle2 = random_spot_on_cake()
    cut = random_spot_on_cake()
    if between(cut, candle1, candle2):
        successes += 1
print(f'empirical probability: {successes}/{trials} = {100 * successes / trials}%')
```
